### **1.1. Sezar Algoritması**

Türk Alfabesinde 29 harf vardır. Her harfi sıra ile 29 kalan sınıfına atıyoruz. <br>

<br>

Bu değerlere p dersek her bir harfi 29 un kalan sınıfında da 3 ötelemek için
\begin{align}
f(p)=(p+3) \mathbf{mod} 29
\end{align} 
dönüşümü ile şifrelenecek metnin sayısal karşılığından şifrelenmiş sayı dizisi oluşturulur. Bu sayı dizisi yukarıdaki tabloya göre tekrar karakter dizisine dönüştürülerek metin şifrelenir.

```python
turkAlfabesi='ABCÇDEFGĞHIİJKLMNOÖPRSŞTUÜVYZ'
```

```python
def karakter_sayi_donusturucu(gelenHarf):
  p = 0
  
  while gelenHarf != turkAlfabesi[p]:
    p = p + 1

  return p
```

```python
karakter_sayi_donusturucu('M')
```

```python
def sayi_karakter_donusturucu(sayi):
  return turkAlfabesi[sayi]
```

```python
sayi_karakter_donusturucu(14)
```

```python
def metin_sayidizisi_donusturucu(metin):
  sayiDizisi = []
  for i in metin:
    sayiDizisi.append(karakter_sayi_donusturucu(i))
  return sayiDizisi
```

```python
metin_sayidizisi_donusturucu('MANİSA')
```

```python
metin_sayidizisi_donusturucu('ŞGTÖZG')
```

```python
def sayidizisi_metin_donusturucu(sayiDizisi):
  metin = ''

  for i in sayiDizisi:
    metin += sayi_karakter_donusturucu(i)

  return metin
```

```python
sayidizisi_metin_donusturucu([15, 0, 23, 5, 15, 0, 23, 11, 13])
```

```python
sayidizisi_metin_donusturucu([15, 0, 16, 11, 21, 0])
```

```python
def sezar_algoritmasi(metin):
  sifreliSayiDizisi = []
  
  for i in metin_sayidizisi_donusturucu(metin):
    sifreliSayiDizisi.append((i + 3) % len(turkAlfabesi))

  return sayidizisi_metin_donusturucu(sifreliSayiDizisi)
```

```python
sezar_algoritmasi('MUSTAFA')
```

### **1.2. Sezar Algoritması Kripto Analizi**

\begin{align}
f^{-1}(p)=(p-3) mod 29
\end{align} 
dönüşümü ile şifreli metni çözümlenir.

```python
def sezar_algoritma_cozucu(sifreliMetin):
  cozulmusSayiDizisi = []
  
  for i in metin_sayidizisi_donusturucu(sifreliMetin):
    cozulmusSayiDizisi.append((i - 3) % len(turkAlfabesi))

  return sayidizisi_metin_donusturucu(cozulmusSayiDizisi)
```

```python
sezar_algoritma_cozucu('ÖYUVÇHÇ')
```

### **1.3. Öteleme Algoritması**

Türk Alfabesinde 29 harf vardır. Her harfi sıra ile 29 kalan sınıfına atıyoruz. <br>

Harf | Sayı
--- | ---
A  | 0
B  | 1
C  | 2
Ç  | 3
D  | 4
... | ...
V  | 26
Y  | 27
Z  | 28

<br>
Bu değerlere p dersek her bir harfi 29 un kalan sınıfında da k ötelemek için
\begin{align}
f(p)=(p+k) \mathbf{mod} 29
\end{align} 
dönüşümü ile şifrelenecek metnin sayısal karşılığından şifrelenmiş sayı dizisi oluşturulur. Bu sayı dizisi yukarıdaki tabloya göre tekrar karakter dizisine dönüştürülerek metin şifrelenir. Burada k anahtar değeridir ve $0<k<29$ olmalıdır.


```python
def k_oteleme_algoritmasi(metin,k):
  sifreliSayiDizisi = []

  for i in metin_sayidizisi_donusturucu(metin):
    sifreliSayiDizisi.append((i + k) % len(turkAlfabesi))

  return sayidizisi_metin_donusturucu(sifreliSayiDizisi)
```

```python
k_oteleme_algoritmasi('MANİSA', 7)
```

### **1.4. Öteleme Algoritması Kripto Analizi**

$n$ kalan sınıfında ($n$ karakterli metin dizisi kullanıldığında),<br> 

$k$ anahtarı için öteleme allgoritması ile şifrelenmiş metni
\begin{align}
f^{-1}(p)=(p-k) \mathbf{mod} n
\end{align} 
dönüşümü ile çözümleriz.

```python
def k_oteleme_algoritma_cozucu(sifreliMetin, k):
  cozulmusSayiDizisi = []

  
  for i in metin_sayidizisi_donusturucu(sifreliMetin):
    cozulmusSayiDizisi.append((i - k) % len(turkAlfabesi))

  return sayidizisi_metin_donusturucu(cozulmusSayiDizisi)
```

```python
k_oteleme_algoritma_cozucu('ŞGTÖZG', 7)
```

### **1.5. Afin Dönüşüm Algoritması**

n kalan sınıfında $(a,b)$ ikili anahtarı ile afin dönüşümü
\begin{align}
f(p)=(ap+b) \mathbf{mod} n
\end{align} 
şeklinde tanımlanır. Burada $a$ ve $n$ aralarında asal olmalıdır.

```python
def afin_algoritmasi(metin, a, b):
  sifreliSayiDizisi = []
  
  for p in metin_sayidizisi_donusturucu(metin):
    sifreliSayiDizisi.append((a * p + b) % len(turkAlfabesi))

  return sayidizisi_metin_donusturucu(sifreliSayiDizisi)
```

```python
afin_algoritmasi('MATEMATİK',2,7)
```

```python
def modul_al(m, n, k): # (m/n) % k 
  i = 0
  while int((m + i * k) / n) !=  (m + i * k) / n :
    i = i + 1
  return  int(((m + i * k) / n) % k) 
```

```python
modul_al(17,2,29)
```

```python
def afin_algoritmasi_cozucu(sifreliMetin, a, b):
  sayiDizisi = []

  for p in metin_sayidizisi_donusturucu(sifreliMetin):
    sayiDizisi.append( modul_al(p - b, a, len(turkAlfabesi))) 

  return sayidizisi_metin_donusturucu(sayiDizisi)
```
